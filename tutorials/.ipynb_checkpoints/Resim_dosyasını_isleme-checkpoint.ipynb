## Kütüphane : Pillow

Kullanacağınız environvent'i kontrol edin. Eğer 'Pillow' isimli kütüphane kurulu değilse, aşağıdaki komutlarla kurabilirsiniz.

### Kütüphaneyi import ederken:

```python
from PIL import Image
```

 ### Resmi açma:

```python
resim = Image.open('image.jpeg', 'r')

resim
```

### Resmin boyutu:

Resmin boyutu, pikselleri değiştirdikten sonra yeni resim yaratırken gerekli bir argüman.

```python
resim_boyutu = resim.size

resim_boyutu
```

### Resmin piksellerine ulaşma:

```python
piksel_listesi = list(resim.getdata())

piksel_listesi[:20]  # Tüm sayfayı kaplamaması için sadece 20 pikselin yapısını gösteriyorum.
```

```python
len(piksel_listesi)
```

### Piksel yapısı:

```python
ilk_piksel = piksel_listesi[0]

ilk_piksel
```

Pikseller, 'red', 'green' ve 'blue' değerleri "integer" olmak üzere, üç elemanlı "tuple" halinde bir "liste" içerisinde listelenmiş.
